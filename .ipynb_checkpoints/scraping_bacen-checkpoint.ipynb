```python
import pandas as pd
import numpy as np
import glob
from datetime import date
from selenium.webdriver import Firefox
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
from time import sleep

# Local de download dos arquivos
path = '/home/caio/Downloads/STP*.csv'

ipca = 433
igpm = 189
selic = 4390
dolar_americano = 3695
dolar_americano_mensal = 3696

# Script
for item in [ipca, igpm, selic, dolar_americano, dolar_americano_mensal]:
    
    # Criando o navegador
    wd = Firefox()
    sleep(2)
    
    # Abrindo o Site
    wd.get('https://www3.bcb.gov.br/sgspub/localizarseries/localizarSeries.do?method=prepararTelaLocalizarSeries')
    sleep(2)
    
    # Saindo da caixa de confirmacao
    try:
        wd.refresh()
    except:
        pass
    
    # Selecionando e informando codigo de busca
    sleep(2)
    wd.find_element(By.ID, 'txCodigo').send_keys(item)
    wd.find_element(By.ID, 'txCodigo').send_keys(Keys.ENTER)
    sleep(2)
    
    # Clicando na caixa de selecao
    wd.find_elements(By.TAG_NAME, 'input')[18].click()
    
    # Clicando no botao
    wd.find_element(By.XPATH, '/html/body/form/center/span/center/table/tbody/tr/td[4]/div/input').click()
    sleep(2)
    
    # Acrescentando data inicial da pesquisa
    wd.find_element(By.NAME, 'dataInicio').click()
    wd.find_element(By.NAME, 'dataInicio').send_keys(Keys.END)
    for i in range(20):
        ActionChains(wd).key_down(Keys.BACKSPACE).perform()
    sleep(2)
    wd.find_element(By.NAME, 'dataInicio').send_keys('01071994')
    
    # Clicando no botao de consulta
    wd.find_elements(By.TAG_NAME, 'input')[8].click()
    sleep(2)

    # Fazendo download do arquivo
    wd.find_elements(By.TAG_NAME, 'a')[10].click()
    sleep(2)
    
    # Fechando o navegador
    wd.close()
```

```python
# Abrindo os arquivos e salvando em dicionario
data = {}
for arquivo in glob.glob(path):
    if '433' in pd.read_csv(arquivo, sep=';', encoding='latin1').columns[1]:
        data['ipca_mes'] = pd.read_csv(arquivo, sep=';', encoding='latin1')[:-1]
    elif '189' in pd.read_csv(arquivo, sep=';', encoding='latin1').columns[1]:
        data['igpm_mes'] = pd.read_csv(arquivo, sep=';', encoding='latin1')[:-1]
    elif '4390' in pd.read_csv(arquivo, sep=';', encoding='latin1').columns[1]:
        data['selic_mes'] = pd.read_csv(arquivo, sep=';', encoding='latin1')[:-1]
    elif  '3695' in pd.read_csv(arquivo, sep=';', encoding='latin1').columns[1]:
        data['dolar_ptax_mes'] = pd.read_csv(arquivo, sep=';', encoding='latin1')[:-1]
    else:
        data['dolar_ptax_mes_fim_periodo'] = pd.read_csv(arquivo, sep=';', encoding='latin1')[:-1]
```

```python
# Tratamento de dados
indices = ['ipca_mes', 'igpm_mes', 'selic_mes']
for key in data.keys():
    # Dividindo taxas por 100, exceto de dolar
    if key in indices:
        data[key][key] = data[key][data[key].columns[1]].apply(lambda x: float(x.replace(',', '.'))/100)
    else:
        data[key][key] = data[key][data[key].columns[1]].apply(lambda x: float(x.replace(',', '.')))
    
    # Adequando o formato de periodo
    data[key]['Data'] = data[key]['Data'].apply(lambda x: ''.join([x.split('/')[1], x.split('/')[0]]))

    if f'{date.today().year}{"%02d" % (date.today().month - 1)}' in list(data[key].Data):
        pass
    else:
        data[key] = data[key].append(pd.DataFrame([[f'{date.today().year}{"%02d" % (date.today().month - 1)}', np.NaN, '0.0000']], columns=data[key].columns)).reset_index().drop(columns='index')

    if f'{date.today().year}{"%02d" % (date.today().month)}' in list(data[key].Data):
        pass
    else:
        data[key] = data[key].append(pd.DataFrame([[f'{date.today().year}{"%02d" % (date.today().month)}', np.NaN, '0.0000']], columns=data[key].columns)).reset_index().drop(columns='index')
    data[key] = data[key].append(pd.DataFrame([[f'{date.today().year}{"%02d" % (date.today().month + 1)}', np.NaN, '0.0000']], columns=data[key].columns)).reset_index().drop(columns='index')

    data[key].rename(columns={'Data': 'periodo'}, inplace=True)

    data[key] = data[key].iloc[:, [0, 2]]
```

```python
# Unindo todos os dados no mesmo data frame
count = 0
for key in data.keys():

    if count == 0:
        df = data[key]
    else:
        df = df.merge(data[key], how='left', on='periodo')

    count += 1
```

```python
# Criando colunas de acumulados
df['ipca_acum'] = [np.NaN if ((float(df.ipca_mes[i]) != 0) | (int(df.periodo[i]) < 202305)) else 1.0 for i in range(len(df))]
df['igp_m_acum'] = [np.NaN if ((float(df.igpm_mes[i]) != 0) | (int(df.periodo[i]) < 202305)) else 1.0 for i in range(len(df))]
df['selic_acum'] = [np.NaN if ((float(df.selic_mes[i]) != 0) | (int(df.periodo[i]) < 202305)) else 1.0 for i in range(len(df))]
```

```python
# Convertendo para floar e limitando quantidade de casas decimais
df['ipca_mes'] = df['ipca_mes'].astype('float64').round(4)
df['igpm_mes'] = df['igpm_mes'].astype('float64').round(10)
df['selic_mes'] = df['selic_mes'].astype('float64').round(4)
df['dolar_ptax_mes'] = df['dolar_ptax_mes'].astype('float64').round(4)
df['dolar_ptax_mes_fim_periodo'] = df['dolar_ptax_mes_fim_periodo'].astype('float64').round(4)
```

```python
# Reordenando os indices pelo periodo decrescente
df = df.sort_values('periodo', ascending=False).reset_index().drop(columns='index')
```

```python
# Preenchendo as colunas de acumulados
for i in range(len(df)):
    if i == 0:
        pass
    else:
        df.ipca_acum[i] =  df.ipca_acum[i-1] * (1 + df.ipca_mes[i])
        df.igp_m_acum[i] =  df.igp_m_acum[i-1] * (1 + df.igpm_mes[i])
        df.selic_acum[i] =  df.selic_acum[i-1] * (1 + df.selic_mes[i])
```

```python
# Reordenando as colunas
df = df[['periodo', 'selic_mes', 'selic_acum', 'ipca_mes', 'ipca_acum', 'igpm_mes', 'igp_m_acum', 'dolar_ptax_mes',
       'dolar_ptax_mes_fim_periodo']]
```

```python
# Adequando a quantidade de casas decimais
def format_number(x):
    return f"{x:.10f}"

df = df.assign(
    igpm_mes=df["igpm_mes"].apply(format_number),
    selic_acum=df["selic_acum"].apply(format_number),
    ipca_acum=df["ipca_acum"].apply(format_number),
    igp_m_acum=df["igp_m_acum"].apply(format_number),
)
```
